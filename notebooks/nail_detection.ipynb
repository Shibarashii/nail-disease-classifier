```python
from pathlib import Path
import sys

ROOT_DIR = Path.cwd().parent
sys.path.insert(0, str(ROOT_DIR))

WEIGHTS_DIR = ROOT_DIR / "outputs" / "yolov8" / "weights" / "best.pt"
```

Initializing YOLO


```python
from ultralytics import YOLO

model = YOLO(WEIGHTS_DIR)
```

## Single Prediction


```python
import matplotlib.pyplot as plt
# Run prediction (disable auto-save here)
IMG_NAME = "1.jpeg"
IMG_PATH = ROOT_DIR / "data" / IMG_NAME

results = model.predict(IMG_PATH, conf=0.1, save=False)

# Show prediction
for r in results:
    im_array = r.plot()  # plot bboxes on image (numpy array BGR)
    plt.imshow(im_array[..., ::-1])  # convert BGR → RGB
    plt.axis("off")
    plt.show()
```

## Live Nail Detection


```python
# Run real-time prediction (0 = default webcam)
model.predict(source=0, show=True, conf=0.8)
```
